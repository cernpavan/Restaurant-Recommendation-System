# Extracting Numeric Ratings from Strings

Sometimes the ratings in a dataset are stored as strings, e.g., "4.5 out of 5" or "3 stars". To convert them into numeric values for analysis:

1. `df['rate'] = df['rate'].str.extract(r'(\d+.\d+|\d+)', expand=True)`  
   - Uses a regular expression to extract the numeric part of the rating from the string.  
   - `\d+.\d+` matches decimal numbers like `4.5`, and `\d+` matches integers like `4`.

2. `df['rate'] = pd.to_numeric(df['rate'], errors='coerce')`  
   - Converts the extracted strings into numeric values (`int` or `float`).  
   - `errors='coerce'` replaces any non-numeric values with `NaN`.


```python
df['rate']=df['rate'].str.extract(r'(\d+.\d+|\d+)',expand=True)
df['rate']=pd.to_numeric(df['rate'],errors='coerce')
```

# Cleaning and Filling Missing Restaurant Types

This section handles the `rest_type` column in the dataset:

- First, it corrects any typos in restaurant types (e.g., changing 'Cafee' to 'Cafe').
- Then, it fills missing values (`NaN`) in `rest_type` based on keywords found in the `listed_in(type)` column:
  - If the type contains 'Delivery', the restaurant type is set as 'Quick Bites'.
  - If it contains 'Dine-out', it is set as 'Casual Dining'.
  - If it contains 'Cafes', it is set as 'Cafe', and so on for other categories like 'Desserts', 'Drinks', 'Buffet', and 'Pubs and bars'.
- `isna()` checks for missing values, and `str.contains(..., na=False)` searches for specific keywords while ignoring missing entries.


```python
df['rest_type']=df['rest_type'].str.replace('Cafee','Cafe')
```

```python
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Delivery', na=False), 'rest_type'] = 'Quick Bites'
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Dine-out', na=False), 'rest_type'] = 'Casual Dining'
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Cafes', na=False), 'rest_type'] = 'Cafe'
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Desserts', na=False), 'rest_type'] = 'Dessert Parlor'
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Drinks', na=False), 'rest_type'] = 'Pub,Bar'
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Buffet', na=False), 'rest_type'] = 'Buffet'
df.loc[df['rest_type'].isna() & df['listed_in(type)'].str.contains('Pubs and bars', na=False), 'rest_type'] = 'Pub,Bar'
```

# Cleaning and Filling Missing Approximate Cost

This code handles the `approx_cost(for two people)` column in the dataset:

- First, it removes commas from the cost values to make them numeric-friendly.
- Then, it converts the column from string to numeric using `pd.to_numeric()`. Any non-numeric entries are converted to `NaN` with `errors='coerce'`.
- After that, it fills missing values (`NaN`) with the **median cost** calculated within specific groups:
  - First by `listed_in(city)` and `rest_type` to account for different cuisines in each city.
  - Then by `listed_in(city)` and `listed_in(type)` to further refine the imputation based on the restaurant type category.
- Using `groupby().transform(lambda x: x.fillna(x.median()))` ensures that each missing value is replaced with the median of its group.


```python
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','')
df['approx_cost(for two people)']=pd.to_numeric(df['approx_cost(for two people)'],errors='coerce')
```

```python
df['approx_cost(for two people)'] = df.groupby(['listed_in(city)', 'rest_type'])['approx_cost(for two people)'].transform(lambda x: x.fillna(x.median()))
df['approx_cost(for two people)'] = df.groupby(['listed_in(city)', 'listed_in(type)'])['approx_cost(for two people)'].transform(lambda x: x.fillna(x.median()))
```

```python
df.head()
```

# Filling Missing Cuisines

This code handles missing values in the `cuisines` column:

- It fills missing entries (`NaN`) using the **most common cuisine** (mode) within specific groups:
  - First grouped by `listed_in(city)` and `rest_type` to capture popular cuisines for each type of restaurant in a city.
  - Then grouped by `listed_in(city)` and `listed_in(type)` to further refine based on the restaurant category.
- `x.mode()[0] if not x.mode().empty else np.nan` ensures that the mode is used if it exists; otherwise, the missing value remains `NaN`.
- Using `groupby().transform()` applies this imputation for each row within its group.


```python
df['cuisines'] = df['cuisines'].fillna(df.groupby(['listed_in(city)', 'rest_type'])['cuisines'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan)) 
df['cuisines'] = df['cuisines'].fillna(df.groupby(['listed_in(city)', 'listed_in(type)'])['cuisines'].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan))
```

```python
df.head()
```

# Filling Missing Ratings

This code handles missing values in the `rate` column:

- First, missing ratings (`NaN`) are filled with the **median rating** of restaurants grouped by `listed_in(city)`, `cuisines`, and `rest_type`.  
  - This ensures that the imputed value reflects similar restaurants within the same city and cuisine type.

- Next, any remaining missing ratings are filled with the **median rating** of all restaurants in the same city.  
  - This serves as a fallback when the more specific group does not provide a value.

- Using `groupby().transform(lambda x: x.fillna(x.median()))` ensures that each missing rating is replaced appropriately within its group.


```python
df['rate'] = df.groupby(['listed_in(city)', 'cuisines','rest_type'])['rate'].transform(lambda x: x.fillna(x.median()))
df['rate'] = df.groupby(['listed_in(city)'])['rate'].transform(lambda x: x.fillna(x.median()))
```
